## 4. Mine agent behavior

Once you have successfully placed an order, it's time to analyze the agents' behaviors and understand the underlying processes. The following code snippet reads the traces generated from your interactions with the coffee shop and creates a basic event log. This event log contains all your past interactions with the shop and can be utilized for Process Mining using SAP Signavio Process Intelligence.
One conversation can consist of multiple traces, as each trace comprises one input message provided by you and the subsequent actions taken by the agents. In other words, each message you send to the coffee shop creates a new trace.

```python
from src import TraceProcessor

processor = TraceProcessor()
processor.process_all_traces(export_as_json=False) # Set `export_as_json` to True to export as JSON instead of CSV
```

In order to upload the event log to SAP Signavio Process Intelligence
1. Locate or create a process in Process Intelligence
2. Click on "Settings" (gear icon) at the top right
3. Locate the generated event log (csv file) in the newly created `generated_event_log` folder.
4. Drag and drop this event log file to the "Step 1: Import data" field, as shown in the image below.
5. At the first import, Process Intelligence will ask you to configure the types of the columns. For this, use the following setting:
    - For the "Case ID", choose the `case_id` column.
    - For the "Activity name", choose the `concept:instance` column.
    - For the "End timestamp", choose the `time:timestamp` column.
    - Then, configure the remaining columns as follows:
      - `identity:id` as type "Text"
      - `concept:name` as type "Choice"
      - `org:resource` as type "Choice"
      - `time_finished` as type "Date time"
      - `duration` as type "Number"
      - `model` as type "Choice"
      - `input_tokens` as type "Number"
      - `response_tokens` as type "Number"
      - `message` as type "Text
      - `tool`as type "Choice"
    - Confirm the mapping by clicking the "Import" button at the bottom right.
6. For subsequent imports, it is sufficient to simply confirm the mapping.
